```python
from PIL import Image
from torchvision import transforms
import torch
import torch.nn as nn
import math
```

```python
input_image = transforms.ToTensor()(Image.open("../data/DRIVE/training/1st_manual/21_manual1.gif").convert('RGB')).unsqueeze(0)
```

1. 转换到ycbcr空间

```python
ycbcr_weight = torch.tensor(
    [[[0.2126, 0.7152, 0.0722]],       # Y
     [[-0.1146, -0.3854, 0.5000]],     # Cb
     [[0.5000, -0.4542, -0.0458]]]    # Cr
).view(3, 3, 1, 1)
ycbcr = nn.functional.conv2d(input_image,ycbcr_weight)
y_in = ycbcr[:, 0:1, :, :]
cb_in = ycbcr[:, 1:2, :, :]
cr_in = ycbcr[:, 2:3, :, :]
```

2. 计算几何属性

$ YCC_{chroma} = \sqrt{Cb^2 + Cr^2} $

$ YCC_{hue angle} = arctan(Cr/Cb) $


```python
y_255 = y_in * 255.0
cb_255 = cb_in * 255.0
cr_255 = cr_in * 255.0
chroma = torch.sqrt(cb_255**2 + cr_255**2)

hue_rad = torch.atan2(cr_255, cb_255)
hue_deg = (torch.degrees(hue_rad) + 360.0) % 360.0
```
