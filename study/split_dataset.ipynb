1. 为验证集、测试集、训练集划分索引

```python
from dlzx import torch as dlzx
```

```python
root_dir = "../data/COVID-CT"

dataset = dlzx.MyImageDataset(root_dir)
```

```python
import random

train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1
seed = 42

assert abs(train_ratio + val_ratio + test_ratio - 1.0) < 1e-6
```

```python
indices = list(range(len(dataset)))
random.Random(seed).shuffle(indices) #固定随机种子
```

```python
n_train = int(train_ratio * len(dataset))
n_val = int(val_ratio * len(dataset))
```

```python
print(n_train,n_val)
```

```python
train_idx = indices[:n_train]
val_idx = indices[n_train:n_train+n_val]
test_idx = indices[n_train + n_val:]
```

2. 用得到的索引，读取数据集

```python
trainsets = []

for idx in train_idx:
    img, label = dataset[idx]
    trainsets.append((img, label))
```

```python
print(trainsets[899])
print(len(trainsets))
```

```python
valsets = []

for idx in val_idx:
    img, label = dataset[idx]
    valsets.append((img, label))
```

```python
print(valsets[10])
print(len(valsets))
```

```python
testsets = []

for idx in test_idx:
    img, label = dataset[idx]
    testsets.append((img, label))

print(testsets[24])
print(len(testsets))
```
